```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
import seaborn as sb
%matplotlib inline
os.chdir('/Users/mars/Downloads/Local/')
df = pd.read_csv('autompg.csv', header=None, delim_whitespace=True)
```

I have imported the libraries needed and separated the data into columns.<br>
Next, I will name the columns to make it easier.

```python
print(df)
df.columns=['mpg', 'cylinders', 'displacement', 'horsepower', 'weight', 'acceleration', 'year', 'origin', 'model name']
```

Now I will replace the "?" with 0, and change datatypes. I also replaced the origins with their proper names.


```python
df['horsepower'] = df['horsepower'].replace(['?'],'0')
df[["mpg", "cylinders", "displacement", "horsepower", "weight", "acceleration"]] = df[["mpg", "cylinders", "displacement", "horsepower", "weight", "acceleration"]].apply(pd.to_numeric, errors='coerce') 

df['origin'] = df.origin.astype(str)
df['origin'] = df['origin'].replace(['1','2','3'],['USA','Asia','Europe'])
```
