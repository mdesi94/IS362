<h4>Dataset #3:</h4>

The final dataset is a bit more complex, and has several NaN values.
Here, we will imagine that we want to compare the numeric data from each year only. The null values for string data are not very important in this situation.

```python
ag.head(10)
```

In this dataset, complete colums with all null values have no use to us. We can <b>get rid of the columns with more than 10 NaN values </b> and have a much better looking dataset. 

```python
ag.dropna(thresh=10, axis = 1)
```

So now, we have all the numeric data for the year columns, and we revoved the values with more than 10 null values that we didn't need.
