<h4> First step is to import packages we will be using, and connect to PostgreSql to read the data.</h4>

```python
import sqlalchemy as db
import pandas as pd

#Connect to postgresql
engine= sqlalchemy.create_engine('postgresql://postgres:postgres123@localhost/chinook')

#check that our database is connected by printing table names
print(engine.table_names())
```

<h4>Next, we will perform queries to select the data we are looking for. </h4>

```python
#write sql query to join tables

sqlQuery = sqlalchemy.text('SELECT "Album"."Title", "Track"."Name", "Customer"."FirstName", "Customer"."LastName"'
'FROM "Album"'
'INNER JOIN "Track" ON "Album"."AlbumId" = "Track"."AlbumId"'
'INNER JOIN "InvoiceLine" ON "Track"."TrackId" = "InvoiceLine"."TrackId"'
'INNER JOIN "Invoice" ON "InvoiceLine"."InvoiceId" = "Invoice"."InvoiceId"'
'INNER JOIN "Customer" ON "Invoice"."CustomerId" = "Customer"."CustomerId"'
'ORDER BY "LastName", "FirstName";')

con = engine

#execute above sql query
result = engine.execute(sqlQuery)
df = result.fetchall()

for row in df:
    print(row)
    
#close the database connection
con.close()
```

<h4>Last, we will print the results as a dataframe.</h4>

```python
df = pd.read_sql(sqlQuery, con)
print(df.head(5))
```
